```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_groups = school_data_complete.groupby("school_name") # Collection by SCHOOL
grade_groups = school_data_complete.groupby("grade") # Collection by GRADE
gender_groups = school_data_complete.groupby("gender") # Collection by GENDER

students_list = school_data_complete["student_name"] # List of all Students
```

```python
# 1. Number of Schools
num_schools = school_groups.size().size
num_schools
```

```python
# 2. Number of Students
num_students = students_list.size
num_students
```

```python
# 3. Total Budget
tot_budget = school_groups.first()["budget"].sum()
tot_budget
```

```python
# 4. Average Math Score
math_avg = school_data_complete["math_score"].mean()
math_avg
```

```python
# 5. Average Reading Score
read_avg = school_data_complete["reading_score"].mean()
read_avg
```

```python
# 6. Overall Passing Score
avg_score = sum([math_avg,read_avg])/2
avg_score
```

```python
# 7. Percent Passing Math
def passing(grade):
    if (grade>=70):
        return 1
    else:
        return 0

math_pass = school_data_complete["math_score"].agg(passing).sum()
math_pass = math_pass/num_students
math_pass
```

```python
# 8. Percent Passing Reading
read_pass = school_data_complete['reading_score'].agg(passing).sum()
read_pass = read_pass/num_students
read_pass
```

```python
# 9. DataFrame Creation
district_summary = pd.DataFrame({'Number of Schools':[num_schools],
                                 'Number of Students':[num_students],
                                 'Total Budget':[tot_budget],
                                 'Average Math Score':[math_avg],
                                 'Average Reading Score':[read_avg],
                                 'Average Passing':[avg_score],
                                 'Percent Passing Math':[math_pass],
                                 'Percent Passing Reading':[read_pass]})

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# For each school in school_group

s_summary = {"School Name":[],
                  "School Type": [],
                  'Total Students': [],
                  'Total School Budget': [],
                  'Per Student Budget': [],
                  'Average Math Score': [],
                  'Average Reading Score': [],
                  'Percent Passing Math': [],
                  'Percent Passing Reading': [],
                  'Overall Passing Rate': [] }

for school in school_groups.groups: # Name saved in school
    
    # Logic for each school
    s = school_groups.get_group(school)
    
    # 0. School Name, saved as dict key for ordering
    s_summary['School Name'].append(school)
    
    # 1. School Type
    s_summary['School Type'].append(s["type"].iat[0])
    
    # 2. Total Students
    stu = s['student_name'].size
    s_summary['Total Students'].append(stu)
    
    # 3. Total School Budget
    b = s['budget'].iat[0]
    s_summary['Total School Budget'].append(b)
    
    # 4. Per Student Budget
    s_summary['Per Student Budget'].append(b/stu)
    
    # 5. Average Math Score
    s_summary['Average Math Score'].append(s['math_score'].mean())
    
    # 6. Average Reading Score
    s_summary['Average Reading Score'].append(s['reading_score'].mean())
    
    # 7. Passing Math
    m = s['math_score'].agg(passing).sum()
    pm = m/stu
    s_summary['Percent Passing Math'].append(pm)
    
    # 8. Passing Reading
    r = s['reading_score'].agg(passing).sum()
    pr = r/stu
    s_summary['Percent Passing Reading'].append(pr)
    
    # 9. Average Passing Rate (note different than above)
    s_summary['Overall Passing Rate'].append(sum([pm,pr])/2)
    
s_summary=pd.DataFrame(s_summary).set_index("School Name")
s_summary
    
```
